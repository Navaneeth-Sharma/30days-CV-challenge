```python
from PIL import Image
import numpy as np
```

```python
# get the channels or see which channels are present
img.getbands()
```

#### Modes
Mode refers to the type and depth of a pixel in an image. Some of the modes that are currently supported are: 

- L for black and white 
- RGB for true color 
- RGBA for true color with transparency mask 
- YCbCr for color video format

```python
img.mode
```

#### Loading and saving 

```python
import base64

with open("../images/dog.png", "rb") as image:
    image_string = base64.b64encode(image.read())
```

```python
# croping images
img = Image.open("../images/dog.png")
box = (100, 50, 650, 650)
cropped_image = img.crop(box)
cropped_image
```
